```python
#Importing Libraries
import numpy as np # To get the data for our plots
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#linspace is a function in numpy
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
#cos wave
plt.plot(x,z)
plt.show()
```

```python
#Multiple Plots:
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,z,'g-*')
plt.subplot(2,2,3)
plt.plot(x,y,'b+-')
plt.subplot(2,2,4)
plt.plot(x,z,'y-*')
plt.savefig('Multiple plot.jpg')
plt.show()
```

```python
#adding title to x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Wave')
plt.title('Sine Curve')
plt.show()
```

```python
#plt.plot(x,y,)

#Multiple graphs in same plot
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.savefig('SinCos.png')      #To download the figure
plt.show()
```

```python
#Bar Graph
fig=plt.figure()
ax=fig.add_axes([0,0,0.5,0.5])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('No. of People')
plt.show()
plt.savefig('BarGraph.jpeg')
```

```python
#Histograms
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
#Box plot -- It is used to plot percentiles
data=[np.random.normal(0,std,100) for std in range(1,4)]
#Rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

```python
#Box plot -- It is used to plot percentiles
data=[np.random.normal(0,std,100) for std in range(1,4)]
#Rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False)
plt.show()
```
